```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Read the image
image = cv2.imread('alpacino.jpg')
```

```python
# Read the grayscale image
gray_image = cv2.imread('alpacino.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Separate the channels
b, g, r = cv2.split(image)
```

```python
# Convert to binary using a threshold of 160
_, binary_image = cv2.threshold(gray_image, 160, 255, cv2.THRESH_BINARY)
```

```python
# Calculate and display the histogram of the original grayscale image
hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
plt.plot(hist)
plt.title('Histogram of Original Grayscale Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Apply histogram equalization to improve brightness
equalized_image = cv2.equalizeHist(gray_image)
```

```python
# Calculate and display the histogram of the equalized image
equalized_hist = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])
plt.plot(equalized_hist)
plt.title('Histogram of Equalized Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Save the processed images
output_folder = os.path.join(os.path.expanduser("."), "MidTerm-2024")
os.makedirs(output_folder, exist_ok=True)
cv2.imwrite(os.path.join(output_folder, "original_gray_image.jpg"), gray_image)
cv2.imwrite(os.path.join(output_folder, "equalized_image.jpg"), equalized_image)
cv2.imwrite(os.path.join(output_folder, "binary_image.jpg"), binary_image)
cv2.imwrite(os.path.join(output_folder, "blue_channel.jpg"), b)
cv2.imwrite(os.path.join(output_folder, "green_channel.jpg"), g)
cv2.imwrite(os.path.join(output_folder, "red_channel.jpg"), r)
```

```python
# Display the original and equalized images
# Show the images
cv2.imshow('Original Image', image)
cv2.imshow('Grayscale Image', gray_image)
cv2.imshow('Binary Image', binary_image)
cv2.imshow('Blue Channel', b)
cv2.imshow('Green Channel', g)
cv2.imshow('Red Channel', r)
cv2.imshow('Equalized Image', equalized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
